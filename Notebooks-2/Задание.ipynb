### Расстояние между двумя точками в пространстве произвольной мерности

```python
def p(A, B):
    if (len(A) == len(B)):
        return np.sqrt(np.sum(np.square(np.sum([A, -B], axis=0))))
    else:
        raise ValueError("Incorrect vectors: ", A, B)
```

### Вспомогательная функция отрисовки изображения

```python
def reimage(data, centers, width=20, height=10):
    new_image = []
    for cluster in data:
        new_image.append(centers[cluster])
    new_image = np.array(new_image).astype(np.uint8)
    new_image = new_image.reshape((image.shape[0],image.shape[1],3))
    plt.figure(figsize=(width, height))
    plt.axis("off")
    plt.imshow(new_image)
    plt.show()
```

### K-Means

```python
class KMeans:
    
    def __init__(self, n_clusters=4, low=0, high=1, max_iterations=10, eps=1, rst=0):
        self.n_clusters = n_clusters
        self.cluster_centers = None
        self.low = low
        self.high = high
        self.max_iterations = max_iterations
        self.eps = eps
        self.rst = rst
        
    def fit(self, data):
        
        ran = np.random.RandomState(self.rst)
        self.cluster_centers = ran.uniform(self.low, self.high, self.n_clusters*data.shape[1])
        self.cluster_centers = self.cluster_centers.reshape((self.n_clusters, data.shape[1]))
            
        u = np.array([0 for x in range(data.shape[0])])
        y = np.zeros((data.shape[0],))
        
        work = True
        count = 0
        
        while (work):
            
            work = False
            
            for i in range(data.shape[0]):
                pos_xi = np.zeros((self.n_clusters,))
                for j in range(self.n_clusters):
                    pos_xi[j] = p(data[i], self.cluster_centers[j])
                y_i = np.argmin(pos_xi)
                u[i] = y_i
                if (np.absolute(y[i] - pos_xi[y_i]) > self.eps):
                    work = True
                y[i] = pos_xi[y_i]

            for k in range(self.n_clusters):
                for dim in range(data.shape[1]):
                    sd = 0
                    cn = 0
                    for i in range(data.shape[0]):
                        if (u[i] == k):
                            sd += data[i][dim]
                            cn += 1
                    if (cn != 0):
                        self.cluster_centers[k][dim] = sd / cn
                        
            count += 1
            
            if (count >= self.max_iterations):
                work = False
                
        return self
    
    def predict(self, data):
        u = np.array([0 for x in range(data.shape[0])])
        for i in range(data.shape[0]):
            pos_xi = np.zeros((self.n_clusters,))
            for j in range(self.n_clusters):
                pos_xi[j] = p(data[i], self.cluster_centers[j])
            y_i = np.argmin(pos_xi)
            u[i] = y_i
        return u
```

```python
proc_data = km.predict(img_data)
```

### Mini Batch K-Means

```python
class MiniBatchKMeans:
    
    def __init__(self, n_clusters=4, low=0, high=1, batch_size=100, max_iterations=10, eps=1, rst=0):
        self.n_clusters = n_clusters
        self.cluster_centers = None
        self.low = low
        self.high = high
        self.batch_size = batch_size
        self.max_iterations = max_iterations
        self.eps = eps
        self.rst = rst
        
    def fit(self, data):
        
        ran = np.random.RandomState(self.rst)
        self.cluster_centers = ran.uniform(self.low, self.high, self.n_clusters*data.shape[1])
        self.cluster_centers = self.cluster_centers.reshape((self.n_clusters, data.shape[1]))
            
        u = np.array([0 for x in range(data.shape[0])])
        y = np.zeros((data.shape[0],))
        
        work = True
        count = 0
        
        while (work):
            
            work = False
            
            batch_ids = ran.choice(range(data.shape[0]), self.batch_size, replace=False)
            batch = np.zeros((self.batch_size, data.shape[1]))
            for i in range(self.batch_size):
                batch[i] = data[batch_ids[i]]
            
            for i in range(batch.shape[0]):
                pos_xi = np.zeros((self.n_clusters,))
                for j in range(self.n_clusters):
                    pos_xi[j] = p(batch[i], self.cluster_centers[j])
                y_i = np.argmin(pos_xi)
                u[i] = y_i
                if (np.absolute(y[i] - pos_xi[y_i]) > self.eps):
                    work = True
                y[i] = pos_xi[y_i]

            for k in range(self.n_clusters):
                for dim in range(batch.shape[1]):
                    sd = 0
                    cn = 0
                    for i in range(batch.shape[0]):
                        if (u[i] == k):
                            sd += batch[i][dim]
                            cn += 1
                    if (cn != 0):
                        self.cluster_centers[k][dim] = sd / cn
                        
            count += 1
            
            if (count >= self.max_iterations):
                work = False
                
        return self
    
    def predict(self, data):
        u = np.array([0 for x in range(data.shape[0])])
        for i in range(data.shape[0]):
            pos_xi = np.zeros((self.n_clusters,))
            for j in range(self.n_clusters):
                pos_xi[j] = p(data[i], self.cluster_centers[j])
            y_i = np.argmin(pos_xi)
            u[i] = y_i
        return u
```

### K-Means (наглядная демонстрация работы алгоритма)

```python
class SpecialKMeans:
    
    def __init__(self, n_clusters=4, low=0, high=1, max_iterations=10, eps=1, rst=0):
        self.n_clusters = n_clusters
        self.cluster_centers = None
        self.low = low
        self.high = high
        self.max_iterations = max_iterations
        self.eps = eps
        self.rst = rst
        
    def fit(self, data):
        
        ran = np.random.RandomState(self.rst)
        self.cluster_centers = ran.uniform(self.low, self.high, self.n_clusters*data.shape[1])
        self.cluster_centers = self.cluster_centers.reshape((self.n_clusters, data.shape[1]))
        
        plt.figure(1)
        plt.plot(*zip(*data), marker='o', color='gray', ls='')
        plt.plot(*zip(*self.cluster_centers), marker='x', color='g', ls='')
        plt.show()
        
        u = np.array([0 for x in range(data.shape[0])])
        y = np.zeros((data.shape[0],))
        
        work = True
        count = 0
        
        while (work):
            
            work = False
            
            for i in range(data.shape[0]):
                pos_xi = np.zeros((self.n_clusters,))
                for j in range(self.n_clusters):
                    pos_xi[j] = p(data[i], self.cluster_centers[j])
                y_i = np.argmin(pos_xi)
                u[i] = y_i
                if (np.absolute(y[i] - pos_xi[y_i]) > self.eps):
                    work = True
                y[i] = pos_xi[y_i]

            for k in range(self.n_clusters):
                for dim in range(data.shape[1]):
                    sd = 0
                    cn = 0
                    for i in range(data.shape[0]):
                        if (u[i] == k):
                            sd += data[i][dim]
                            cn += 1
                    if (cn != 0):
                        self.cluster_centers[k][dim] = sd / cn
                        
            count += 1
            
            if (count >= self.max_iterations):
                work = False
                
            plt.figure(1)
            for i in range(data.shape[0]):
                if (u[i]%7 == 0):
                    clr = 'red'
                if (u[i]%7 == 1):
                    clr = 'green'
                if (u[i]%7 == 2):
                    clr = 'blue'
                if (u[i]%7 == 3):
                    clr = 'orange'
                if (u[i]%7 == 4):
                    clr = 'lightgreen'
                if (u[i]%7 == 5):
                    clr = 'lightblue'
                if (u[i]%7 == 6):
                    clr = 'lightred'
                plt.plot(*zip(data[i]), marker='o', color=clr, ls='')
            plt.plot(*zip(*self.cluster_centers), marker='x', color='black', ls='')
            plt.show()
                
        return self
```
